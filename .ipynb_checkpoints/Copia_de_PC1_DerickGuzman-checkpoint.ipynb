**1.Keep all the columns but Ranking:**

**2. Find the country with the minimum CO2 emission in the world.**



```python
minco2_country = carbon_copy.loc[carbon_copy['metric_tonnes_of_CO2'].idxmin(), 'name']
print("\n2. Country with the lowest CO2 emissions in the world:", minco2_country)
```

**3. Find the country with the minimum CO2 emission per Region.**

```python
min_by_region = carbon_copy.loc[carbon_copy.groupby('region')['metric_tonnes_of_CO2'].idxmin(), ['region', 'name', 'metric_tonnes_of_CO2']]
min_by_region
```

**4. Create a new column by keeping the square root of the the original CO2 column.**

```python
carbon_copy['CO2_square'] = carbon_copy['metric_tonnes_of_CO2']**0.5
carbon_copy
```

**5. Compute the average of CO2. Then, create a new column. This new column is the original CO2 minus the average computed.**

```python
mean_co2 = carbon_copy['metric_tonnes_of_CO2'].mean()
carbon_copy['CO2_minus_average'] = carbon_copy['metric_tonnes_of_CO2'] - mean_co2
carbon_copy
```
